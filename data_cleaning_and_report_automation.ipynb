```python
import pandas as pd
```

```python
data = pd.read_csv('/Mall_Customers.csv')
```

```python
print(data)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python


# Check missing values
print("Missing Values:")
print(data.isnull().sum())


```

```python
# Remove duplicates
data = data.drop_duplicates()
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Generate report
with open("report.txt", "w") as report:
    report.write("MALL CUSTOMERS REPORT\n")
    report.write("="*30 + "\n\n")

    report.write("Dataset Shape:\n")
    report.write(str(data.shape))
    report.write("\n\n")

    report.write("Missing Values:\n")
    report.write(str(data.isnull().sum()))
    report.write("\n\n")

    report.write("Summary Statistics:\n")
    report.write(str(data.describe()))

print("Report Generated Successfully!")
```
