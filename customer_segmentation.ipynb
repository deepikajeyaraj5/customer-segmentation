**7.KMeans Clustering**

```python
data.columns
```

```python
x=data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
from sklearn.cluster import KMeans
```

```python
k_means=KMeans()
k_means.fit(x)
```

```python
k_means=KMeans(n_clusters=5)
k_means.fit_predict(x)
```

**8.Elbow Method To Find Optimal Number Of Clusters**

```python
wcss=[]
for i in range(1,11):
  k_means=KMeans(n_clusters=i)
  k_means.fit(x)
  wcss.append(k_means.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss)
plt.title("Elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()
```

**9.Model Training**

```python
x=data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
k_means = KMeans(n_clusters=5,random_state=42)
y_means = k_means.fit_predict(x)
```

```python
y_means
```

```python
plt.scatter(x.iloc[y_means==0,0],x.iloc[y_means==0,1],s=30,c='red',label="cluster 1")
plt.scatter(x.iloc[y_means==1,0],x.iloc[y_means==1,1],s=30,c='yellow',label="cluster 2")
plt.scatter(x.iloc[y_means==2,0],x.iloc[y_means==2,1],s=30,c='black',label="cluster 3")
plt.scatter(x.iloc[y_means==3,0],x.iloc[y_means==3,1],s=30,c='blue',label="cluster 4")
plt.scatter(x.iloc[y_means==4,0],x.iloc[y_means==4,1],s=30,c='green',label="cluster 5")
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1],s=30,c='magenta')
plt.title("customer segmentation")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

**10.Save The Model**

```python
import joblib
```

```python
joblib.dump(k_means, "customer_segmentation.pkl")
```

```python
model = joblib.load("customer_segmentation.pkl")
```

```python
import ipywidgets as widgets
from IPython.display import display

income = widgets.IntText(
    value=50,
    description='Income'
)

score = widgets.IntText(
    value=50,
    description='Score'
)

button = widgets.Button(
    description='Predict'
)

output = widgets.Output()

def predict_customer(b):

    with output:

        output.clear_output()

        result = model.predict(
            [[income.value,
              score.value]]
        )

        cluster = result[0]

        print("Cluster Number =", cluster)

        if cluster == 0:
            print("Medium Income - Medium Spending")

        elif cluster == 1:
            print("High Income - Low Spending")

        elif cluster == 2:
            print("Low Income - Low Spending")

        elif cluster == 3:
            print("Low Income - High Spending")

        else:
            print("High Income - High Spending")

button.on_click(predict_customer)

display(income)
display(score)
display(button)
display(output)
```
